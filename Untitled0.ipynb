```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60]
y = [ 1, 2, 3, 4,5,6]
plt.plot(x,y)
plt.xlabel('Ages')
plt.ylabel('Medium Salary (USD)')
plt.title('Medium Salary (USD) by Age')
plt.show()
```
